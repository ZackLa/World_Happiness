## These missing regions for three different files, which I went back to csv to check, they didn't name the country consistent, and when I first doing the dictionary pairs with country and regions, there are some missing country, that's why I have missing values for region, so I have to rename country, and luckily I was able to find all of them except for gambia, which mean I have to look for myself online to check its region.

```python
#Getting the right country that's formatted in csv file
t1 = d2[d2['Country'].isin(['Trinidad and Tobago', 'Belize', 
                            'North Cyprus', 'Somalia', 'Namibia',
                            'South Sudan', 'Taiwan', 'Hong Kong','Macedonia', 'Gambia'])]
t1
```

```python
#Rename Country
d3.iat[32,1] = 'Taiwan'
d3.iat[70,1] = 'Hong Kong'
d4.iat[37,1] = 'Trinidad and Tobago'
d4.iat[57,1] = 'North Cyprus'
d5.iat[38,1] = 'Trinidad and Tobago'
d5.iat[63,1] = 'North Cyprus'
d5.iat[83,1] = 'Macedonia'
```

```python
for i, k in enumerate(t1.index):
    print(i,k)
```

```python
country_region1 = {}
# Dict variable to pair country corresponding region
for i, k in enumerate(t1.index):
    country_region1[t1.loc[k]['Country']] = t1.loc[k]['Region']
country_region1['Gambia'] = 'Sub-Saharan Africa'
```

```python
country_region1
```

```python
#Based on its country it is going to look for the value in dict in country_region1 to get the corresponding region
d3.loc[d3['Region'].isnull(),"Region"] = d3["Country"].apply(lambda x: country_region1.get(x))
d4.loc[d4['Region'].isnull(),"Region"] = d4["Country"].apply(lambda x: country_region1.get(x))
d5.loc[d5['Region'].isnull(),"Region"] = d5["Country"].apply(lambda x: country_region1.get(x))
```

## Target null value for TrustCorruption

```python
# Decide to replace the mean value to the only null value
d4['TrustCorruption'] = d4.TrustCorruption.fillna(d4['TrustCorruption'].mean())
```
